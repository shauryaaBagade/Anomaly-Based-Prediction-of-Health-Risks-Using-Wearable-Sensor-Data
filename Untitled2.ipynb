```python
import pandas as pd
import os

file_name = "train_FD001.txt"

if os.path.exists(file_name):
    train = pd.read_csv(file_name, sep=r"\s+", header=None)
    print(f"Successfully loaded {file_name}.")
    print(train.head())
else:
    print(f"Error: The file '{file_name}' was not found.")
    print("Please ensure you have uploaded 'train_FD001.txt' using the 'files.upload()' cell.")
    print("If you have already uploaded it, please try re-running the upload cell and then this cell.")
```

```python
train.shape
```

```python
# Define column names
columns = ['engine_id', 'cycle',
           'op_setting1', 'op_setting2', 'op_setting3']

sensor_cols = [f'sensor_{i}' for i in range(1, 22)]


new_column_names = columns + sensor_cols # This list has 26 elements


if train.shape[1] > len(new_column_names):
    print(f"Warning: DataFrame has {train.shape[1]} columns, expected {len(new_column_names)}. Trimming to {len(new_column_names)} columns.")
    train = train.iloc[:, :len(new_column_names)]
elif train.shape[1] < len(new_column_names):
    print(f"Error: DataFrame has {train.shape[1]} columns, expected {len(new_column_names)}. Cannot proceed with renaming.")
    # This case is less likely given the original error message, but included for completeness.
    raise ValueError("DataFrame has fewer columns than expected for renaming.")

# Now assign the column names
train.columns = new_column_names

train.head()
```

```python
# Find maximum cycle for each engine
rul = train.groupby('engine_id')['cycle'].max().reset_index()
rul.columns = ['engine_id', 'max_cycle']

# Merge with original dataset
train = train.merge(rul, on='engine_id')

# Calculate RUL
train['RUL'] = train['max_cycle'] - train['cycle']
train['RUL'] = train['RUL'].clip(upper=125)

# Drop max_cycle column
train = train.drop('max_cycle', axis=1)

train[['engine_id', 'cycle', 'RUL']].head()
```

```python
from sklearn.feature_selection import VarianceThreshold

# Select only sensor columns
sensor_cols = [col for col in train.columns if 'sensor' in col]

X_sensors = train[sensor_cols]

# Apply variance threshold
selector = VarianceThreshold(threshold=0.0001)
X_selected = selector.fit_transform(X_sensors)

# Get selected sensor names
selected_sensors = X_sensors.columns[selector.get_support()]

print("Selected Sensors:")
print(selected_sensors)
```

```python
train = train[['engine_id', 'cycle', 'RUL'] + list(selected_sensors)]

train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train[selected_sensors] = scaler.fit_transform(train[selected_sensors])

train.head()
```

```python
import numpy as np

def create_sequences(df, window_size, feature_cols):
    sequences = []
    labels = []

    for engine in df['engine_id'].unique():
        engine_data = df[df['engine_id'] == engine]

        for i in range(len(engine_data) - window_size):
            seq = engine_data[feature_cols].iloc[i:i+window_size].values
            label = engine_data['RUL'].iloc[i+window_size]

            sequences.append(seq)
            labels.append(label)

    return np.array(sequences), np.array(labels)

window_size = 30

X_train, y_train = create_sequences(train, window_size, selected_sensors)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=42
)

print("Training shape:", X_tr.shape)
print("Validation shape:", X_val.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential()

model.add(LSTM(64, return_sequences=True, input_shape=(30, 14)))
model.add(Dropout(0.3))

model.add(LSTM(32))
model.add(Dropout(0.3))

model.add(Dense(16, activation='relu'))
model.add(Dense(1))

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    X_tr, y_tr,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=64,
    callbacks=[early_stop]
)
```

```python
def create_test_sequences(df, window_size, feature_cols):
    sequences = []

    for engine in df['engine_id'].unique():
        engine_data = df[df['engine_id'] == engine]

        # Take last window_size cycles
        seq = engine_data[feature_cols].iloc[-window_size:].values

        sequences.append(seq)

    return np.array(sequences)

X_test = create_test_sequences(test, window_size, selected_sensors)

print("X_test shape:", X_test.shape)

```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Load true RUL values
rul_test = pd.read_csv("RUL_FD001.txt", header=None)
y_test_true = rul_test.values.flatten()

# Calculate RMSE
rmse_test = np.sqrt(mean_squared_error(y_test_true, y_test_pred))

print("Final Test RMSE:", rmse_test)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(y_test_true, label='Actual RUL')
plt.plot(y_test_pred, label='Predicted RUL')
plt.legend()
plt.title("Test Set: Actual vs Predicted RUL")
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
import numpy as np

anomaly_threshold = 30
y_test_anomaly = (y_test_true <= anomaly_threshold).astype(int)
y_test_pred_flat = y_test_pred.flatten()
lstm_scores = -y_test_pred_flat

best_f1_lstm, best_thresh_lstm, best_p_lstm, best_r_lstm = 0, 0, 0, 0

for thresh in np.percentile(lstm_scores, np.linspace(50, 99, 200)):
    preds = (lstm_scores > thresh).astype(int)
    f1 = f1_score(y_test_anomaly, preds, zero_division=0)
    if f1 > best_f1_lstm:
        best_f1_lstm, best_thresh_lstm = f1, thresh
        best_p_lstm = precision_score(y_test_anomaly, preds, zero_division=0)
        best_r_lstm = recall_score(y_test_anomaly, preds, zero_division=0)

lstm_anomaly_preds = (lstm_scores > best_thresh_lstm).astype(int)
auc_lstm = roc_auc_score(y_test_anomaly, lstm_scores)

print("LSTM Results (Optimized):")
print(f"Precision : {best_p_lstm:.4f}")
print(f"Recall    : {best_r_lstm:.4f}")
print(f"F1        : {best_f1_lstm:.4f}")
print(f"AUC       : {auc_lstm:.4f}")
```

```python
from sklearn.ensemble import IsolationForest

X_test_last  = X_test[:, -1, :]
X_train_last = X_train[:, -1, :]

y_train_anomaly = (y_train <= anomaly_threshold).astype(int)
actual_contamination = y_train_anomaly.sum() / len(y_train_anomaly)
print(f"Contamination rate from train: {actual_contamination:.4f}")

iso_forest = IsolationForest(contamination=actual_contamination, random_state=42)
iso_forest.fit(X_train_last)
if_scores = -iso_forest.score_samples(X_test_last)

best_f1_if, best_thresh_if, best_p_if, best_r_if = 0, 0, 0, 0

for thresh in np.percentile(if_scores, np.linspace(50, 99, 200)):
    preds = (if_scores > thresh).astype(int)
    f1 = f1_score(y_test_anomaly, preds, zero_division=0)
    if f1 > best_f1_if:
        best_f1_if, best_thresh_if = f1, thresh
        best_p_if = precision_score(y_test_anomaly, preds, zero_division=0)
        best_r_if = recall_score(y_test_anomaly, preds, zero_division=0)

if_anomaly_preds = (if_scores > best_thresh_if).astype(int)
auc_if = roc_auc_score(y_test_anomaly, if_scores)

print("\nIsolation Forest Results (Optimized):")
print(f"Precision : {best_p_if:.4f}")
print(f"Recall    : {best_r_if:.4f}")
print(f"F1        : {best_f1_if:.4f}")
print(f"AUC       : {auc_if:.4f}")
```

```python
lstm_norm = (lstm_scores - lstm_scores.min()) / (lstm_scores.max() - lstm_scores.min())
if_norm   = (if_scores - if_scores.min()) / (if_scores.max() - if_scores.min())

best_f1_h, best_alpha, best_thresh_h, best_p_h, best_r_h = 0, 0, 0, 0, 0

for alpha in np.linspace(0, 1, 21):
    hybrid_scores = alpha * lstm_norm + (1 - alpha) * if_norm
    for thresh in np.percentile(hybrid_scores, np.linspace(50, 99, 100)):
        preds = (hybrid_scores > thresh).astype(int)
        f1 = f1_score(y_test_anomaly, preds, zero_division=0)
        if f1 > best_f1_h:
            best_f1_h, best_alpha, best_thresh_h = f1, alpha, thresh
            best_p_h = precision_score(y_test_anomaly, preds, zero_division=0)
            best_r_h = recall_score(y_test_anomaly, preds, zero_division=0)

hybrid_scores = best_alpha * lstm_norm + (1 - best_alpha) * if_norm
hybrid_preds  = (hybrid_scores > best_thresh_h).astype(int)
auc_hybrid    = roc_auc_score(y_test_anomaly, hybrid_scores)

print("Hybrid Fusion Results:")
print(f"LSTM weight : {best_alpha:.2f}  |  IF weight: {1-best_alpha:.2f}")
print(f"Precision   : {best_p_h:.4f}")
print(f"Recall      : {best_r_h:.4f}")
print(f"F1          : {best_f1_h:.4f}")
print(f"AUC         : {auc_hybrid:.4f}")
```
